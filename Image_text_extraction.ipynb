# Text Extraction & Detection From Image

```python
#function to detect text from image
def extract_text_from_image(image_path) :
    
    import cv2
    import easyocr
    import matplotlib.pyplot as plt
    
    image = cv2.imread(image_path)
    reader = easyocr.Reader(['en'])
    text_data = reader.readtext(image)
    
    for i in range(len(text_data)) :
        coordinates = text_data[i][0]
        x_min, y_min = [int(min(val)) for val in zip(*coordinates)]
        x_max, y_max = [int(max(val)) for val in zip(*coordinates)]
        text_detected_img = cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)
        
    plt.imshow(cv2.cvtColor(text_detected_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
```

```python
#calling the function
extract_text_from_image("magazine.jpg")
```

```python
extract_text_from_image("text.png")
```
